```python
import pandas as pd
```

les données sont regroupées en 60 fichiers CSV (un fichier par mois)

```python
import os
arr = os.listdir('./data')
print(arr, len(arr))
```

```python
DF_list = list()

for doc in arr:
    DF_list.append(pd.read_csv('./data/' + doc))
```

```python
for df in DF_list:
    df.dropna()
```

une incompatibilité des types de certaines colones a rendu le merge des dataframes impossibles 
pour corriger cette incompatibilité :

* identifier et merge les dataframes dont les colonnes ont le meme type 
    
* identifier les colonnes avec un type différent et l'adapter pour pouvoir merge toutes le données

On remarque dans un premier temps que la colonne goal est soit de type int soit de type float 
on passe goal en float dans tous les cas 


```python
data_1, data_2, data_3, data_4 = list(), list(), list(), list()
for i in range(len(DF_list)):
    if(DF_list[i].dtypes.value_counts()[0] == 21):
        if(DF_list[i].dtypes.value_counts()[1] == 8):
            data_1.append(DF_list[i])
        else:
            data_2.append(DF_list[i])
    elif(DF_list[i].dtypes.value_counts()[1] == 8):
        data_3.append(DF_list[i])
    else:
        data_4.append(DF_list[i])
    
```

```python
data1 = data_1[0]
for i in range(1, len(data_1)) :
    #print(data1[i].dtypes)
    data1 = pd.merge(data1, data_1[i], how='outer')
data1.info()
```

```python
data2 = data_2[0]
for i in range(1, len(data_2)) :
    #print(data1[i].dtypes)
    data2 = pd.merge(data2, data_2[i], how='outer')
data2.info()
```

```python
data3 = data_3[0]
for i in range(1, len(data_3)) :
    #print(data_3[i].dtypes.value_counts())
    data3 = pd.merge(data3, data_3[i], how='outer')
data3.info()
```

```python
data4 = data_4[0]
for i in range(1, len(data_4)) :
    #print(data_3[i].dtypes.value_counts())
    data4 = pd.merge(data4, data_4[i], how='outer')
data4.info()
```

```python
data1 = data1.astype({"goal": float})
data = pd.merge(data1, data2, how='outer')
data.info()
```

```python
data3 = data3.astype({"goal": float})
df = pd.merge(data3, data4, how='outer')
df.info()
```

```python
print(data.dtypes)
print(df.dtypes)
```

Les colonnes 'friends', 'is_backing', 'is_starred', 'permissions' sont souvent nulles 
et donc interprétés comme float quand dans un dataframe il n'y a que des na
on supprime ces colonnes 
On peut merge tous les dataframes

```python
print(df.shape)
print(df['permissions'].isna().sum())
print(data.shape)
print(data['permissions'].isna().sum())
df.drop(columns=['friends', 'is_backing', 'is_starred', 'permissions'])
data['friends'] = data['friends'].fillna(0)
data['is_backing'] = data['is_backing'].fillna(0)
data['is_starred'] = data['is_starred'].fillna(0)
data['is_starred'] = data['is_starred'].fillna(0)
data['permissions'] = data['permissions'].fillna(0)
```

```python
data = pd.merge(data, df, how='outer')
data.info()
```

```python
del data1, data2, data3, data4, data_1, data_2, data_3, data_4, DF_list
```

```python
data = data.drop_duplicates()
```
